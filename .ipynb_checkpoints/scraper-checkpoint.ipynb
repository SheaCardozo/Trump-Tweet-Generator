```python
import requests

CLIENT_ACCESS_TOKEN = "Izal14EhsrfmAGUbiAUqCEtZWn6Tr6F9sD2mKAzoxAiAAkc1cIgY4dhn9f35yiGz"

YEEZUS = 72

BASE_URI = "https://api.genius.com"

def _get(path, params=None, headers=None):

    url = '/'.join([BASE_URI, path])

    token = "Bearer {}".format(CLIENT_ACCESS_TOKEN)

    if headers:
        headers['Authorization'] = token
    else:
        headers = {"Authorization": token}

    response = requests.get(url=url, params=params, headers=headers)
    response.raise_for_status()

    return response.json()

def get_artist_songs(artist_id):

    current_page = 1
    next_page = True
    songs = []

    while next_page:

        path = "artists/{}/songs/".format(artist_id)
        params = {'page': current_page}
        data = _get(path=path, params=params)

        page_songs = data['response']['songs']

        if page_songs:
            songs += page_songs
            current_page += 1
        else:
            next_page = False

    return songs
```

```python
from bs4 import BeautifulSoup

def scrape_lyrics(url):

    response = requests.get(url)
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')
    
    lyrics = ''
    
    for div in soup.findAll('div', attrs = {'class': 'lyrics'}):
        lyrics = lyrics + div.text

    return lyrics
```

```python
lyriclist = []
urllist = []
titlelist = []
filter = ['Dates','Speech','Adrenalin (Kırıldı Karyola)','баунд 2 бурундуки','Interview','5 Best Beats','Cinematic','Booklet','Compilation','Behind the Scenes','Tracklist','Setlist','Rant','*','Film','🐉','Artwork','Credits','Skit','Notepad','[','Art', 'Lecture','Cover','On ', 'Kanye', 'Making of']
for song in songs:
    skip = False
    url = song['url']
    title = song['title']
    for i in filter:
        if i in title:
            skip = True
    if not skip:
        lyrics = scrape_lyrics(url).strip()
        lyriclist.append(lyrics)
        urllist.append(url)
        titlelist.append(title)
```
